```python
import numpy as np
```

```python
#READ THE DATASET
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris) #BUNCH TYPE OF OBJECT
```

```python
print(iris.DESCR) #FEATURES - SEPAL LENGTH, SEPAL WIDTH, PETAL LENGTH, PETALWIDTH
```

```python
#GRAB THE FEATURE AND LABEL INFORMATION
X = iris.data
```

```python
X #SEPAL LENGTH, SEPAL WIDTH, PETAL LENGTH, PETAL WIDTH IN NUMPY ARRAY
```

```python
#GRAB LABELS: CLASSIFICATIONS
Y = iris.target
Y
```

```python
#CONVERT INTO ONE-HOT ENCODING SYSTEM FOR KERAS

#VECTOR THAT IS 0 WHEN DOESN'T MATCH UP

#class 0 - - > [1, 0, 0]
#class 1 - - > [0, 1, 0]
#class 2 - - > [0, 0, 1]
```

```python
from keras.utils import to_categorical
```

```python
Y = to_categorical(Y)
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(X, Y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#SCALE ALL TO VALUES 0-1
scaler_object = MinMaxScaler()
```

```python
#FIT TO TRAINING DATA
scaler_object.fit(X_TRAIN)
```

```python
scaled_X_TRAIN = scaler_object.transform(X_TRAIN)
```

```python
scaled_X_TEST = scaler_object.transform(X_TEST)
```

```python
#SCALED VERSIONS OF THE DATA

from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
#ADD A LAYER

#NO OF NEURONS, NO OF FEATURES, ACTIVATION FUNCTION
model.add(Dense(8, input_dim = 4, activation='relu'))
model.add(Dense(8, input_dim = 4, activation='relu'))
model.add(Dense(3, activation='softmax'))

#COMPILE THE MODEL
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
#ASK FOR SUMMARY OF MODEL
model.summary()
```

```python
#FIT THE MODEL ON SCALED DATA

#VERBOSE DEFINES HOW DETAILED PROGRESS IS TO BE SHOWN
model.fit(scaled_X_TRAIN, Y_TRAIN, epochs=150, verbose=1)
```

```python
#95% FINAL ACCURACY
```

```python
#PART - 2


#SCALE THE TEST DATA: ALREADY DONE

np.argmax(model.predict(scaled_X_TEST), axis=-1)
```

```python
#COMPARE PREDICTIONS WITH THE ONE-HOT ENCODED Y_TEST
predictions = np.argmax(model.predict(scaled_X_TEST), axis=-1)
```

```python
#TRANSFORM THE Y_TEST BACK TO CLASSES
Y_TEST = Y_TEST.argmax(axis=1)
```

```python
#COMPARE THE PREDICTIONS WITH TRUE VALUES
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
confusion_matrix(Y_TEST, predictions)
```

```python
print(classification_report(Y_TEST, predictions))
```
