```python
import pandas as pd

npr = pd.read_csv('../TextFiles/npr.csv')

npr.head()
```

```python
#NO LABEL COLUMN TO INDICATE WHICH TOPIC IT BELONGS TO

#TOTAL ROWS
len(npr)
```

```python
#TRY TO USE UNSUPERVISED LEARNING TO FIGURE OUT TOPICS THEY BELONG TO


#PREPROCESSING
from sklearn.feature_extraction.text import CountVectorizer


#IGNORE TERMS PRESENT IN ALMOST ALL DOCUMENTS [HIGH DF]
cv = CountVectorizer(max_df=0.9, min_df=2, stop_words = 'english') 
#DISCARD WORDS SHOWING IN 90% OR MORE DOCUMENTS
#SHOW UP IN ATLEAST 2 DDOCUMENTS
```

```python
#NO TRAIN-TEST SPLIT IN UNSUPERVISED LEARNING


#DOCUMENT TERM MATRIX
dtm = cv.fit_transform(npr['Article'])

dtm
```

```python
#PERFORM LDA ON THE DOCUMENT TERM MATRIX

#IMPORT
from sklearn.decomposition import LatentDirichletAllocation

#INSTANTIATE
LDA = LatentDirichletAllocation(n_components = 7, random_state = 42)
#NUMBER OF TOPICS = N_COMPONENTS
#RANDOM STATE AS WORDS ASSIGNED RANDOMLY IN START


#FIT THE DOOCUMENT TERM MATRIX
LDA.fit(dtm)
```

```python

#GRAB VOCABULARY OF WORDS

#ALL UNIQUE WORDS IN LIST OF DOCUMENTS
len(cv.get_feature_names()) #THIS IS THE LIST OF ALL UNIQUE WORDS IN DOCUMENTS
```

```python
#GRAB SOME RANDOM WORDS
import random

random_word_id = random.randint(0, 54777)

cv.get_feature_names()[random_word_id] #grab a random word
```

```python
#GRAB THE TOPICS
#GET FROM THE TRAINED LDA
len(LDA.components_) #TOTAL NUMBER OF TOPICS
```

```python
#LDA.COMPONENTS IS A NUMPY ARRAY OF WORDS X TOPICS PROB. DISTRIBUTION
LDA.components_
```

```python
single_topic = LDA.components_[0]
```

```python
#RETURNS INDEX POSITIONS THAT WOULD SORT THIS ARRAY
single_topic.argsort() #INDEX POSITIONS OF WORDS WITH HIGHEST PROBABILITY FOR THAT TOPIC


#GRAB THE TOP 10 WORDS FOR THE TOPIC [INDEX POSITIONS]
top_ten_words = single_topic.argsort()[-20:]
```

```python
#CAN FIGURE OUT THE TOPIC FROM THESE WORDS: INSURANCE, HEALTH
for index in top_ten_words:
    print(cv.get_feature_names()[index])
```

```python
#GRAB THE HIGHEST PROBABILITY WORDS PER TOPIC
for index, topic in enumerate(LDA.components_):
    print(f"THE TOP 15 WORDS FOR TOPIC #{index}")
    
    print([cv.get_feature_names()[index] for index in topic.argsort()[-15:]])
    
    print('\n')
    
    print('\n')
```

```python
#CREATE A COLUMN FOR TOPIC NO
topic_results = LDA.transform(dtm)
```

```python

#PROBABILITIES OF EACH DOCUMENT FOR EACH TOPIC
topic_results[0].round(2)
```

```python
#FIND INDEX POSITION OF HIGHEST PROBABILITY
topic_results[0].argmax() #INDEX OF TOPIC WITH MAX PROB FOR FIRST DOCUMENT
```

```python
npr['Topic_number'] = topic_results.argmax(axis=1)
```
